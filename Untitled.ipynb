```python
def address (text) : 
   
    if ("서울" in text) and ("서울특별시" not in text) : 
        text = text.replace("서울", "서울특별시")

    elif ("경기" in text) and ("경기도" not in text) : 
        text = text.replace("경기", "경기도")

    elif ("강원" in text) and ("강원도" not in text) : 
        text = text.replace("강원", "강원도")

    elif ("부산" in text) and ("부산광역시" not in text) : 
        text = text.replace("부산", "부산광역시")

    elif ("대구" in text) and ("대구광역시" not in text) : 
        text = text.replace("대구", "대구광역시")

    elif ("울산" in text) and ("울산광역시" not in text) : 
        text = text.replace("울산", "울산광역시")

    elif ("대전" in text) and ("대전광역시" not in text) : 
        text = text.replace("대전", "대전광역시")

    elif ("제주" in text) and ("제주특별자치도" not in text) : 
        text = text.replace("제주", "제주특별자치도")

    elif ("충북" in text) : 
        text = text.replace("충북", "충청북도")

    elif ("충남" in text) : 
        text = text.replace("충남", "충청남도")

    elif ("경북" in text) : 
        text = text.replace("경북", "경상북도")

    elif ("경남" in text) : 
        text = text.replace("경남", "경상남도")

    elif ("전북" in text) : 
        text = text.replace("전북", "전라북도")

    elif ("전남" in text) : 
        text = text.replace("전남", "전라남도")

    elif ("인천" in text) and ("인천광역시" not in text) : 
        text = text.replace("인천", "인천광역시")    

    elif ("세종" in text) and ("세종특별자치시" not in text) : 
        text = text.replace("세종", "세종특별자치시")

    return text
```

```python
def addsimple (text) :
    text = text.split("로 ")
    
    if len(text) == 2 :
        text = text[0] + "로"
    
    else :
        text = text[0].split("길 ")
        
        if len(text) == 2 :
            text = text[0] + "길"
        
    return text
```

```python
def addward (text) :
    text = text.split(" ")[0]
    for item in ["광역", "특별자치", "특별"] : 
        text = text.replace(item, "")
    return text
```

```python
def esta (text) : 
    
    

    YY = text[:4]
    MM = text[4:6]
    DD = text[6:]
    return f"{YY}-{MM}-{DD}"
```

```python
def YN(text) : 
    if text == "Y" :
        text = True
    elif text == "N" :
        text = False
    return text
```
